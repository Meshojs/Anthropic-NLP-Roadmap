```python
from nltk.stem import PorterStemmer , WordNetLemmatizer # stemming and lemmatizer
from nltk.tokenize import word_tokenize # tokenize word
from nltk.corpus import stopwords # stopwords
import spacy.parts_of_speech as pos

sentence = "I've eaten computer and it was amazing" 
stop_words = stopwords.words("english")
tokenized = word_tokenize(sentence)
lemmatizer = WordNetLemmatizer()
tokenized = [ lemmatizer.lemmatize(word) for word in tokenized if word not in stop_words]
tokenized
```

<code>NER</code>

```python
import spacy 
nlp = spacy.load("en_core_web_sm")
sentence = "Mashary was happy today"
doc = nlp(sentence)
for ner in doc.ents : 
    print(ner.text , " | " ,  ner.label_ ,  " | " , spacy.explain(ner.label_))
```
