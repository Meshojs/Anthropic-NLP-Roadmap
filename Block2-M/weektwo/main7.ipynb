<code>Feature representation : Embedding</code>

```python
import spacy
```

```python
nlp = spacy.load("en_core_web_md")
```

```python
!python -m spacy download en_core_web_md
```

```python
sentences = [
    "machine learning is powerful",
    "I love eating pizza",
    "music is really good !"
]

docs = [nlp(doc) for doc in sentences]


query = "deep learning and neural networks are the future"


query_doc = nlp(query)

similar = [(doc.text, doc.similarity(query_doc)) for doc in docs]

sorted_results = sorted(similar, key=lambda x: x[1], reverse=True)

for text, score in sorted_results:
    print(score, text)
```
