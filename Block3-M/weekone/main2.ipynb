```python
sentence = "I love deep learning and ai"
seq_length = 2

words = sentence.split()
vocab = {word: idx for idx, word in enumerate(set(words))}

x = []
y = []

for i in range(len(words) - seq_length):
    seq = words[i:i+seq_length]       # مثال: ['I']
    target = words[i+seq_length]      # الكلمة اللي بعدها

    # نحول كل كلمة في seq ل index
    x.append([vocab[word] for word in seq])
    y.append(vocab[target])

print("X:", x)
print("Y:", y)
```
