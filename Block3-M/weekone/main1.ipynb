```python
from gensim.models import Word2Vec

sentences = [
    """ 
Artificial intelligence is transforming the way humans interact with technology. It refers to systems that can perform tasks that usually require human intelligence, such as understanding language, recognizing images, and making decisions. Machine learning is a core part of artificial intelligence, where models learn patterns from data instead of being explicitly programmed. Deep learning, a subset of machine learning, uses neural networks with many layers to process complex information.

In recent years, natural language processing has become one of the most important areas of AI. It enables computers to understand and generate human language, powering applications like chatbots, translation systems, and voice assistants. Models based on transformer architectures have significantly improved performance in language understanding tasks.

AI is widely used in industries such as healthcare, finance, and transportation. In healthcare, it helps in diagnosing diseases and analyzing medical images. In finance, it is used for fraud detection and algorithmic trading. Self-driving cars are another example of AI applications, combining computer vision and decision-making systems.

However, artificial intelligence also raises important concerns. Issues like bias, data privacy, and job displacement need to be addressed carefully. Researchers and engineers are working to make AI systems more fair, transparent, and reliable.

The future of AI holds great potential. As technology advances, AI systems are expected to become more efficient and capable, helping solve complex global problems and improving everyday life.
    """
]

from nltk.tokenize import sent_tokenize, word_tokenize

text = sentences[0]

sentences_tokenized = [
    word_tokenize(sent.lower())
    for sent in sent_tokenize(text)
]


sentences_tokenized

model = Word2Vec(
    sentences=sentences_tokenized,
    vector_size=100,
    window=1, # (3~2) understand the meaning words | (4~5) understand the meaning of the general idea
    min_count=1
)


model.wv.most_similar("ai")
```
