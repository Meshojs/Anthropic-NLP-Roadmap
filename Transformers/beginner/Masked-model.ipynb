A clever monkey living by a river outsmarted a greedy crocodile trying to catch him. By pretending the crocodile was a stone, the monkey tricked it into opening its mouth, then hopped on its head to safely cross to an island of fruit, proving that wit beats strength

```python
from transformers import pipeline
from textwrap import fill
from pprint import pprint


mlm = pipeline("fill-mask")

def wrap(doc):
    print(fill(doc , fix_sentence_endings=True ,replace_whitespace=False))

sentence = """
A  <mask> <mask> a river outsmarted a crocodile trying to catch him. By pretending the crocodile was a stone, the monkey tricked it into opening its mouth, then hopped on its head to safely cross to an island of fruit, proving that wit beats strength
"""
# wrap(sentence)
arr = sentence.split()
results = mlm(sentence, top_k=1)
mask_count = 0
# top_k best guess per mask
for i, word in enumerate(arr):
    if word == "<mask>":
        arr[i] = results[mask_count][0]["token_str"]
        mask_count += 1


arr
```
