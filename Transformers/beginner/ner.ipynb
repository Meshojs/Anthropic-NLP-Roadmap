```python
import pickle
from transformers import pipeline
from nltk.tokenize.treebank import TreebankWordDetokenizer
```

```python
# load Train, Test data
with open("ner_train.pkl" , "rb") as f : 
    train_data = pickle.load(f)

with open("ner_test.pkl" , "rb") as f : 
    test_data = pickle.load(f)

inpt = []
target = []
for sentence in test_data : 
    current_sentence = []
    current_target = []
    for token , tag in sentence : 
        current_sentence.append(token)
        current_target.append(tag)
    inpt.append(current_sentence)
    target.append(current_target)    
```

```python
detokenizer = TreebankWordDetokenizer()
```

```python
import torch as t
device = "cuda" if t.cuda.is_available() else "cpu"
print(device)
ner = pipeline("ner" , aggregation_strategy="simple" , device=0)
sent = detokenizer.detokenize(inpt[9])
result = ner(sent)

f=[]
for dict in result:
    if dict["entity_group"] in target[9]:
        f.append(dict["entity_group"])
    else : 
        f.append("O")

f
```
