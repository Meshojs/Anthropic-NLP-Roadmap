```python
from transformers import pipeline
import pandas as pd
from textwrap import fill
```

```python

summarizer = pipeline("summarization") 
df = pd.read_csv("bbc_text_cls.csv")
paragraph = df.iloc[0]["text"]
result = summarizer(paragraph, max_length=130, min_length=30)
summary = result[0]["summary_text"]
print(fill(summary, fix_sentence_endings=True, width=70))
```
