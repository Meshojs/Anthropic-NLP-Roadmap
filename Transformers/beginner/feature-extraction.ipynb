```python
from transformers import pipeline
import pandas as pd 
import numpy as np

df = pd.read_csv("bbc_text_cls.csv")

featurizer = pipeline("feature-extraction")

data = [" ".join(text.split()[:100]) for text in df["text"][:2250]]
e = featurizer(data)


```

```python
embeddings = [i[0][0] for i in e]             # grab [CLS] from each
embeddings = np.array(embeddings) 
embeddings.shape
```

```python
from sklearn.neighbors import NearestNeighbors
from pprint import pprint
model = NearestNeighbors()
model.fit(embeddings)

distances, indices = model.kneighbors(embeddings[2224].reshape(1,-1) , 20)

for _ in indices: 
    for i in _ : 
        print(df["labels"][i])
```
