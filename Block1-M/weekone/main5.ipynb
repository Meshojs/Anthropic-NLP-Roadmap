```python
import numpy as np 
from sklearn.model_selection import train_test_split
```

```python
xtrain = np.array([[1], [2], [3], [4]])
ytrain = np.array([[2], [4], [6], [8]]) 
```

<code>
 Hidden layer:

W1 shape = (in_features, hidden_neurons)

b1 shape = (1, hidden_neurons)

Output layer:

W2 shape = (hidden_neurons, out_features)

b2 shape = (1, out_features)


```python
def relu(x): 
    return np.maximum(x,0)

def relu_derivative(x):
    return (x > 0).astype(float) # 0 and 1


def nerural_network(xtrain ,ytrain , xtest, in_features=1 , out_features=4, hidden_neurons=4): 
    w1 = np.random.rand(in_features , hidden_neurons)
    b1 = np.random.rand(1 , hidden_neurons)
    
    w2 = np.random.rand(hidden_neurons , out_features)
    b2 = np.random.rand(1 , out_features)
    
    epoch = 5000
    
    for i in range(epoch): 
        
        Z1 = xtrain@w1 + b1
        A1 = relu(Z1)
        
        Z2 = A1@w2 + b2
        y_hat = Z2        
        loss = np.mean((ytrain-y_hat)**2)
        
        # Output layer gradients
        dZ2 = 2 * (y_hat - ytrain) / 4
        dW2 = A1.T @ dZ2
        db2 = np.sum(dZ2, axis=0, keepdims=True)

        # Hidden layer gradients
        dA1 = dZ2 @ w2.T
        dZ1 = dA1 * relu_derivative(Z1)
        dW1 = xtrain.T @ dZ1
        db1 = np.sum(dZ1, axis=0, keepdims=True)

        lr=1e-3

        w1 -= lr * dW1
        b1 -= lr * db1
        w2 -= lr * dW2
        b2 -= lr * db2

    
    Z1 = xtest@w1+b1
    A1 = relu(Z1)
    Z2 = A1@w2+b2
    y_pred = Z2
    return y_pred


xtest = np.array([[5],[6],[7],[8]])

np.round(nerural_network(xtrain ,ytrain , xtest, in_features=1 , out_features=1, hidden_neurons=4))
        
```
