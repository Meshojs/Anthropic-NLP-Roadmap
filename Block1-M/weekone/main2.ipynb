```python
import numpy as np 


def add_nan(x , nan_ratio ): 
    """
        Args
            x (array) : Multi-dim array before adding NaN Values.
            nan_ratio : NaN given from user from 0.0 to 1.0 will be in array 
    """
    
    total_values = x.size
    nan_distro = int(total_values * nan_ratio) # 3
    flat_x = x.flatten()
    flat_x[:nan_distro] = np.nan
    x = flat_x.reshape(x.shape)

    return x

def zero_out(x):
    """
        Args 
            x (array) : Multi-dim array before adding NaN Values.
        returns 
            x without zeros 
    """
    return x[x!=0]
    
def clean(x, **mode):
    if mode['mode'] == "zeros" :
        # x.astype(float)
        x = np.where(np.isnan(x) , 0.0 , x)
        return x
        

    
    
def generate_batch(batch_size, features, *corrupt_ops:str, **config:str):
    arr = np.random.rand(batch_size , features)
    

    for i in corrupt_ops : 
        if len(i) != 0 : 
            if i == "nan": 
                print(add_nan(arr , config['nan_ratio']))
            if i == "zero_out":
                print(zero_out(arr))
                 
            return clean(arr , mode="zeros")
            
        else : 
            return "No Corrupt_ops were found"
    

x = generate_batch(
    2, 3,
    "nan", "zero_out",
    nan_ratio=0.5,
    mode="zeros"
)

x
```
