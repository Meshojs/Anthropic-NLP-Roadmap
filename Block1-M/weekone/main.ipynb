```python
import numpy as np 
from time import time
```

```python
def dec(fn):
    def wrapper(*args): 
        s = time( )
        res = fn(*args)
        e = time()
        print("This function took about : " , round(e-s , 4) , " seconds")    
        return res 
    return wrapper
    
```

```python
@dec
def fn(matOne:np.ndarray, matTwo:np.ndarray) -> np.array:
    matOne = matOne * 5
    matTwo = matTwo.T
    
    return matOne@matTwo

matOne = np.array(
    [1 , 2 , 3 , 4]
)
#1x4
matTwo = np.array(
    [1 , 2 , 3 , 4]
)
#1x4

fn(matOne , matTwo)
```
